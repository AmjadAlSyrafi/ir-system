## 1. Load a sample dataset

```python
import ir_datasets

# Load a small BEIR or TREC dataset
# dataset = ir_datasets.load('beir/scifact/test')
# docs = list(dataset.docs_iter())[:500]
# print(f'Loaded {len(docs)} documents')

# Quick demo with synthetic data
sample_docs = [
    {'id': 'doc1', 'content': 'Information retrieval systems use inverted indexes for fast lookup.'},
    {'id': 'doc2', 'content': 'BM25 is a probabilistic ranking function used in search engines.'},
    {'id': 'doc3', 'content': 'Dense retrieval with sentence transformers enables semantic search.'},
    {'id': 'doc4', 'content': 'TF-IDF weighs term frequency against inverse document frequency.'},
    {'id': 'doc5', 'content': 'Hybrid search combines sparse and dense retrieval for better results.'},
]
pd.DataFrame(sample_docs)
```

## 2. Preprocess documents

```python
from preprocessing.preprocessor import TextPreprocessor

prep = TextPreprocessor(lowercase=True, remove_stopwords=True, lemmatization=True)
for doc in sample_docs:
    print(f"{doc['id']}: {prep.preprocess(doc['content'])}")
```

## 3. BM25 Retrieval

```python
from retrieval.models.bm25_model import BM25Model

bm25 = BM25Model()
bm25.fit(sample_docs)

query = 'semantic search dense vectors'
results = bm25.search(query, top_k=5)
pd.DataFrame(results)
```

## 4. TF-IDF Retrieval

```python
from retrieval.models.tfidf_model import TFIDFModel

tfidf = TFIDFModel()
tfidf.fit(sample_docs)
results = tfidf.search(query, top_k=5)
pd.DataFrame(results)
```

## 5. Evaluation Metrics

```python
from evaluation.metrics import evaluate

retrieved = [r['id'] for r in results]
relevant  = ['doc3', 'doc5']

metrics = evaluate(retrieved, relevant, k_values=[1, 3, 5])
pd.Series(metrics).round(4)
```
